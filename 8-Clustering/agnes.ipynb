### Imports


```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

### Data load


```python
# Load data
df = pd.read_csv("data/insta.csv")
X = df[['Instagram visit score', 'Spending_rank(0 to 100)']]

# Compute hierarchical clustering linkage
linkage_matrix = linkage(X, method='ward')
```

### Plot Dendogram


```python

# Plot dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix)
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

### Model Training and Visualization


```python
# Applying Agglomerative Clustering with 3 clusters
hierarchical = AgglomerativeClustering(n_clusters=5)
df['Hierarchical Cluster'] = hierarchical.fit_predict(X)

# Scatter plot
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=df['Hierarchical Cluster'], cmap='viridis', alpha=0.6)
plt.xlabel('Instagram visit score')
plt.ylabel('Spending Rank')
plt.title('Hierarchical Clustering')
plt.show()
```
